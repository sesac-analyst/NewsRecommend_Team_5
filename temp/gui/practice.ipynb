```python
from tkinter import *
```

```python
window= Tk()

window.title("News Recommend")
window.geometry("640x400+100+100")
window.resizable(False, False)

label=Label(window, text="뉴스 URL을 입력하세요.")
label.pack()

# window.iconbitmap("/PYTHONWORKSPACE/Tkinter/xingtu_icon.ico")
e = Entry(window, width=30)
e.pack()
# e.insert(0, "뭔데")

def Click():
    input_str = e.get()
    label = Label(window, text=input_str)
    label.pack()
    # label.destroy()

btn = Button(window, text="Find", command=Click)
btn.pack()

window.mainloop()
```

```python
import tkinter as tk
from tkinter import scrolledtext

# 관련기사1과 관련기사2를 생성하는 함수 (예시 데이터)
def generate_related_articles(news_text):
    # 입력된 뉴스 기사를 기반으로 관련 기사를 생성하는 로직을 여기에 구현할 수 있음.
    # 현재는 예시 데이터를 반환함.
    related_articles1 = [f"관련기사1 - {i+1}: {news_text[:10]} ..." for i in range(5)]
    related_articles2 = [f"관련기사2 - {i+1}: {news_text[:10]} ..." for i in range(5)]
    return related_articles1, related_articles2

# 뉴스 기사 입력 후 버튼 클릭 시 관련 기사를 출력하는 함수
def on_generate_articles():
    news_text = news_entry.get("1.0", tk.END).strip()
    
    # 뉴스 기사를 입력했을 때 관련 기사를 생성
    if news_text:
        related1, related2 = generate_related_articles(news_text)
        
        # 관련기사1과 관련기사2를 각각의 텍스트 위젯에 출력
        related1_text.delete("1.0", tk.END)  # 기존 내용을 지우고 새로 출력
        related2_text.delete("1.0", tk.END)
        
        for article in related1:
            related1_text.insert(tk.END, article + "\n")
        
        for article in related2:
            related2_text.insert(tk.END, article + "\n")
    else:
        # 뉴스 기사가 입력되지 않았을 경우 경고
        related1_text.delete("1.0", tk.END)
        related2_text.delete("1.0", tk.END)
        related1_text.insert(tk.END, "뉴스 기사를 입력하세요.")
        related2_text.insert(tk.END, "뉴스 기사를 입력하세요.")

# 메인 윈도우 생성
window = tk.Tk()
window.title("연관 뉴스 추천")
window.geometry("640x500+100+100")
# window.resizable(False, False)

# 안내
label=Label(window, text="뉴스 URL을 입력하세요.")
label.grid(row=0, column=0, columnspan=2, padx=10, pady=10)

# 뉴스 기사 입력 창 (스크롤 가능하게 설정)
news_entry = scrolledtext.ScrolledText(window, wrap=tk.WORD, width=50, height=10)
news_entry.grid(row=1, column=0, columnspan=2, padx=10, pady=10)

# 버튼: 관련기사 생성
generate_button = tk.Button(window, text="관련기사 생성", command=on_generate_articles)
generate_button.grid(row=2, column=0, columnspan=2, pady=10)

# 관련기사1 표시 창 (왼쪽 아래)
related1_label = tk.Label(window, text="관련기사1")
related1_label.grid(row=3, column=0, padx=10, pady=5, sticky="w")

related1_text = scrolledtext.ScrolledText(window, wrap=tk.WORD, width=40, height=10)
related1_text.grid(row=4, column=0, padx=10, pady=5)

# 관련기사2 표시 창 (오른쪽 아래)
related2_label = tk.Label(window, text="관련기사2")
related2_label.grid(row=3, column=1, padx=10, pady=5, sticky="w")

related2_text = scrolledtext.ScrolledText(window, wrap=tk.WORD, width=40, height=10)
related2_text.grid(row=4, column=1, padx=10, pady=5)

# 메인 루프 시작
window.mainloop()
```
