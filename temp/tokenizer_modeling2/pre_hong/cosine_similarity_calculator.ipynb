```python
import pandas as pd
import numpy as np
from joblib import Parallel, delayed
```

```python
data_table = pd.read_csv("C:/Users/SesacPython/Desktop/dataset/뉴스추천시스템/vector_3months_data.csv")
```

```python
def str2vec(x):
    return np.fromstring(' '.join(x.replace('\n', '').split())[1:-1].strip(), sep=' ')

vec_list = Parallel(n_jobs=-1)(delayed(str2vec)(i) for i in data_table.vector)
vec_matrix = np.stack(vec_list)
vec_matrix_norm = vec_matrix / np.linalg.norm(vec_matrix, axis=1, keepdims=True)
```

```python
def output_with_cos(url):
    input_vector = str2vec(data_table[data_table['article_url']==url]['vector'].iloc[0])
    input_norm = input_vector / np.linalg.norm(input_vector)
    cos_similarity_list = np.dot(vec_matrix_norm, input_norm)
    data_table['cos_similarity'] = cos_similarity_list
    n_top5 = list(data_table[data_table['tag'].str.startswith('n')].nlargest(5,'cos_similarity')['article_url'])
    d_top5 = list(data_table[data_table['tag'].str.startswith('d')].nlargest(5,'cos_similarity')['article_url'])

    return n_top5, d_top5
```

```python
output_with_cos(input("기사 url을 입력하세요: "))
```
