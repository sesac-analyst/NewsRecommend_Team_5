```python
import pandas as pd
import re
```

```python
open_path = './news_data.csv'

df = pd.read_csv(open_path)
```

```python
copy_df = df.copy()
copy_df.info()
```

```python
copy_df.drop_duplicates(subset='url', keep='first', inplace=True)
copy_df.drop_duplicates(subset='title', keep='first', inplace=True)
copy_df.drop_duplicates(subset='content', keep='first', inplace=True)
copy_df.dropna(subset=['content'], inplace=True)
```

```python
copy_df.info()
```

```python
copy_df.head(10)
```

```python
# 문장내 <>, [] 구문 제거
copy_df['clear_str'] = copy_df['content'].apply(lambda x: re.sub(r'<.*?>', '', x))
copy_df['clear_str'] = copy_df['clear_str'].apply(lambda x: re.sub(r'\[.*?\]', '', x))
```

```python
copy_df[['content', 'clear_str']].head(5)
```

```python
# 단어별 처리 함수
def process_words(words, del_words, stopwords):
        processed_words=[]
        for word in words:
            if word == '기자':
                if processed_words:
                    processed_words.pop()
                continue
            
            if any(del_word in word for del_word in del_words):
                continue

            if '@' in word:
                continue
            # word = re.sub(r'\d+','', word)
            word = re.sub(r'[^\w\s.]','', word)
            
            if re.search('[a-zA-Z]', word):
                word = word.lower()
            
            for stopword in stopwords:
                if word.endswith(stopword):
                    word = word[:-len(stopword)]
            
            if len(word) <= 1:
                continue
                    
            if word:
                processed_words.append(word)
        return processed_words
```

```python
# 정리 단어들 목록
press_name = copy_df['press'].unique()
stopword = ['기자', 'com', '.co', '저작권', '무단', '전재', '재배포', 'Copyr', 'copyr', '경향비즈'
        '영상', '취재', '편집', '문의', '금지', '특파원', '아이뉴스', '한경', '뉴스', '보도합니다']
```

```python
# 단어로 정리
copy_df['split_word'] = copy_df['clear_str'].str.split()
copy_df['processed_words'] = copy_df['split_word'].apply(lambda words: process_words(words, press_name, stopword))
```

```python
copy_df[['content', 'processed_words']].head(5)
```

```python
# 문장 만들기
copy_df['clear_sentence'] = copy_df['processed_words'].apply(lambda words: ' '.join(words))
copy_df['clear_sentence_split'] = copy_df['clear_sentence'].apply(lambda sentence: re.split(r'(?<!\d)\.(?!\d)', sentence))
```

```python
copy_df[['content', 'clear_sentence_split']].head(5)
```

```python
clear_sentence_df = copy_df[['title','category', 'url', 'date', 'press', 'reporter', 'content', 'clear_sentence_split']]
```

```python
save_path = './news_data_clear_sentence.csv'
clear_sentence_df.to_csv(save_path, index=False, encoding='utf-8-sig')
```

### 클렌징 이후

```python
import pandas as pd
import numpy as np
import re
from konlpy.tag import Okt
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import ast
from sklearn.metrics.pairwise import cosine_similarity
```

```python
open_path = './news_data_clear_sentence.csv'
clear_sentence_df = pd.read_csv(open_path)
```

```python
clear_sentence_df.head(3)
```

```python
# 문자열로 저장된 데이터를 실제 리스트로 변환
def convert_to_list(x):
    if isinstance(x, str):  # 데이터가 문자열일 경우만 변환
        return ast.literal_eval(x)
    return x  # 이미 리스트인 경우 변환하지 않음
```

```python
clear_sentence_df['clear_sentence_split'] = clear_sentence_df['clear_sentence_split'].apply(convert_to_list)


# 형태소 분석기 초기화
okt = Okt()

# 형태소 분석 및 TaggedDocument 생성
tagged_data = []
for i, sentences in enumerate(clear_sentence_df['clear_sentence_split'][:100]):
    for sentence in sentences:
        if sentence.strip():  # 빈 문장 방지
            words = okt.morphs(sentence)  # 형태소 분석 수행
            tagged_data.append(TaggedDocument(words=words, tags=[f"SENT_{i}"]))
```

```python
# for doc in tagged_data:
#     print(doc)
```

```python
# Doc2Vec 모델 초기화 (병렬 처리 활성화)
model = Doc2Vec(
    vector_size=50,     # 벡터 크기 (최적화를 위해 크기를 줄임)
    window=5,           # 컨텍스트 윈도우 크기
    min_count=2,        # 최소 단어 빈도 (빈도수가 낮은 단어 무시)
    workers=8,          # 사용하고자 하는 CPU 스레드 수 (최대 CPU 코어 사용)
    epochs=5,          # 에포크 수 
    dm=1                # Distributed Memory 모델 사용
)
```

```python
model.build_vocab(tagged_data)
model.train(tagged_data, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.save("doc2vec_model.model")
```

```python
# 각 문서의 벡터 추출
doc_vectors = [model.dv[tag] for tag in model.dv.index_to_key]

# 모든 문서간의 유사도 계산
similarities = cosine_similarity(doc_vectors)

# 유사도가 높은 상위 10개의 문서 찾기
most_similar_docs_with_similarity = {}
for idx, similarities_row in enumerate(similarities):
    # 자신을 제외한 상위 10개 문서의 인덱스와 유사도 점수
    similar_indices = similarities_row.argsort()[::-1][1:11]
    similar_tags = [model.dv.index_to_key[i] for i in similar_indices]
    similar_scores = [similarities_row[i] for i in similar_indices]
    
    # 유사한 문서 태그와 유사도 점수를 함께 저장
    most_similar_docs_with_similarity[model.dv.index_to_key[idx]] = list(zip(similar_tags, similar_scores))

# 결과 출력
for tag, similar_info in most_similar_docs_with_similarity.items():
    print(f"문서 {tag}와 가장 유사한 상위 10개 문서:")
    for sim_tag, sim_score in similar_info:
        print(f"  태그: {sim_tag}, 유사도: {sim_score:.4f}")
```
